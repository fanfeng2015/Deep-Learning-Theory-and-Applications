```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def plot(n):
    x = np.zeros((n, 1))
    y = np.zeros((n, 1))
    y_corrupted = np.zeros((n, 1))
    for i in range(n):
        x[i, 0] = np.random.uniform(-3, 3)
    for i in range(n):
        y[i, 0] = x[i, 0] ** 2
        y_corrupted[i, 0] = x[i, 0] ** 2 + np.random.normal(0, 1)

    plt.plot(x, y, 'k.')
    plt.plot(x, y_corrupted, 'r.')
    plt.show()
```

```python
plot(300)
```
